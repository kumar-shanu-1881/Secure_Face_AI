```python
import os
import cv2 as cv
import csv
import time
from tqdm import tqdm

from insightface.app import FaceAnalysis

# ----------------------------------------
# Load RetinaFace
# ----------------------------------------

print("Loading RetinaFace...")

app = FaceAnalysis(
    name="buffalo_l",
    allowed_modules=['detection'],
    providers=['CUDAExecutionProvider', 'CPUExecutionProvider']
)

app.prepare(
    ctx_id=0,           # GPU 0
    det_size=(640,640)
)

print("RetinaFace Loaded Successfully!\n")

# ----------------------------------------
# Dataset
# ----------------------------------------

IMAGE_FOLDER="/kaggle/input/datasets/ashwingupta3012/human-faces/Humans"

image_files=[
    f for f in os.listdir(IMAGE_FOLDER)
    if f.lower().endswith((".jpg",".jpeg",".png"))
]

print(f"Found {len(image_files)} images.")

# ----------------------------------------
# CSV
# ----------------------------------------

csv_file="RetinaFace_Evaluation_Results.csv"

total_faces=0
total_latency=0

with open(csv_file,"w",newline="",encoding="utf-8") as file:

    writer=csv.writer(file)

    writer.writerow([
        "Image Name",
        "Faces Detected",
        "Latency (ms)"
    ])

    for img_name in tqdm(image_files):

        img_path=os.path.join(IMAGE_FOLDER,img_name)

        img=cv.imread(img_path)

        if img is None:
            continue

        start=time.perf_counter()

        faces=app.get(img)

        latency=(time.perf_counter()-start)*1000

        face_count=len(faces)

        writer.writerow([
            img_name,
            face_count,
            round(latency,2)
        ])

        total_faces+=face_count
        total_latency+=latency

avg_latency=total_latency/len(image_files)

print("\n========== BENCHMARK SUMMARY ==========")
print("Model            : RetinaFace")
print(f"Images Processed : {len(image_files)}")
print(f"Total Faces      : {total_faces}")
print(f"Average Latency  : {avg_latency:.2f} ms")
print(f"CSV File Saved   : {csv_file}")
print("=======================================")
```
